# Seattle Air Pollution

# Background

In recent years there has been much attention paid to possible health 
risks from very small particles in the air.  These particles are 
collectively referred to as `particulate matter air pollution' (PM).  In 
the northeast US the concentrations of PM are highest in the summer, when 
they include diesel and gasoline exhaust and particles produced by large 
power stations in the industrial midwest.  Many studies have found that 
both cardiovascular and pulmonary morbidity and mortality are very 
slightly higher on days when PM concentrations are high.

In Seattle there is relatively little diesel and almost no coal smoke.  
Concentrations of PM are highest on cold clear winter nights, partly 
because of large amounts of wood smoke and partly because the air is 
stagnant and the pollution is trapped near the surface.

Because of the differences in seasonal pattern and composition of Seattle 
particulate air pollution there is interest in ascertaining whether the 
associations with morbidity and mortality are still found here.  We have 
data on hospital admissions for respiratory disease in King County 
hospitals, and as a control, hospital admissions for gastrointestinal 
disease, which is not thought to be affected by air pollution.

The data set contains variables

| Variable | Description                                             |
|:--------|:-----------------------------------------------------|
| date     | measured in days since 1970-1-1                         |
| mo       | month of the year                                       |
| yr       | two-digit year                                          |
| dow      | day of the week (1=Sunday)                              |
| pm1      | Concentration (micrograms/cubic meter) of particles between about 0.1 and 1 microns in diameter, averaged over the day |
| temp     | maximum temperature in Fahrenheit                       |
| stagno   | Number of hours in the day with low or no wind          |
| admyng   | hospital admissions for respiratory disease in King County hospitals, in people 5-65 years of age |
| gasyng   | hospital admissions for gastrointestinal disease in King County hospitals, in people 5-65 years of age |

# Step 3: Graphs used for this project

1. Bar chart is used to compare the Concentration of particles between 0.1 and 1 microns in diameter, averaged over the day.



```python
df = pd.read_csv(r'http://courses.washington.edu/b517/Datasets/pm.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
# Extract data from day of the week and concentration of particles in the air column
# Create Dataframe
dow_col = df['dow']
pm1_col = df['pm1']

#Plotting Bar chart 
plt.bar(dow_col, pm1_col,color='orange')
plt.xlabel('Day of the Week')
plt.ylabel('Concentration of Particles')
plt.title('Concentration of particles in the air over the week')

plt.show()
```

# Findings and Analysis

The bar chart above compares the variations in air pollution levels on various days of the week. Sunday starts from 1 mentioned in the bar chart above, Monday, Tuesday, Wednesday, Thursday, Friday, and Saturday indicates 2,3,4,5,6, and 7 respectively. It seems that the level of air pollution vary greatly on the day of the week due to different activities such as industrial patterns, traffic activities. As shown in the graph that weekdays have high pollution levels because of increased vehicles on the road and industrial operations. 





```python
#Plotting a histogram to show how many people are suffering from respirtory diseases due to air pollution.

# Create Dataframe

pm1_col = df['pm1']
admyng_col = df['admyng']

#Plotting Histogram 

plt.bar(admyng_col, pm1_col, color='blue')
plt.xlabel('Concentration of Particles in the air over the week')
plt.ylabel('Respiratory disease')
plt.title('Number of patients suffered in Air Pollution')

plt.show()
```

# Findings and Analysis

The histogram above shows the number of people admitted to the hospital for respiratory disease in King County hospitals between the ages of 5 and 65. It indicates that there is a correlation between the concentration of particles in the air and the number of people experiencing respiratory and breathing problems.
